```python
import pandas as pd
df = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vThqKe7eAgIUnm5dGtzqNQ3L3T2CJRxZC7_v6DJ6DZfO78x16D85Scq7UBl7v48NA/pub?gid=1469769408&single=true&output=csv')
df.head(2000)
```

```python
pd.set_option('display.max_columns', None)
df.info()
```

```python
""" Esta tabela mostra a media salarial dos funcionarios demitidos e dos não demitidos por Área de Formação.
Analisando friamente estes dados, a relação do salario medio pela area de formação dos funcionarios deminitos e não demetidos
eles não apresentam uma boa explicação para correlação dos pedidos de demissão """

area_salario = df.groupby(['Demissão','Área de Formação'])[['Salário Mensal']]
area_salario.mean()
```

```python
""" Esta tabela mostra a media salarial dos funcionarios demitidos e dos não demitidos por Estado Civil.
Analisando friamente estes dados, a relação do salario medio pelEstado Civil dos funcionarios deminitos e não demetidos
eles não apresentam uma boa explicação para correlação dos pedidos de demissão """

area_escol = df.groupby(['Demissão','Estado civil'])[['Salário Mensal']]
area_escol.mean()
```

```python
"""Esta tabela mostra o nivel de satisfação do trabalho dos funcionarios demitidos e dos não demitidos por departamento
e a porcentagem de cada departamento por nivel de satisfação com o trabalho. Neste caso podemos observar uma alta porcentagem
do nivel de insatisfação com o trabalho no Departamento de Recursos Humanos, tanto dos demitidos e dos não demitidos, o que
pode indicar uma sobrecarga de trabalho e uma insatisfação geral com a rotina de trabalho."""

dep_satisf = df.groupby(['Demissão','Satisfação com o trabalho','Departamento'])[['Departamento']]
dep_satisf = df.groupby(['Demissão', 'Satisfação com o trabalho', 'Departamento']).size().reset_index(name='Total')

total_por_categoria = dep_satisf.groupby(['Demissão', 'Departamento'])['Total'].transform('sum')
dep_satisf['Porcentagem'] = (dep_satisf['Total'] / total_por_categoria) * 100

display(dep_satisf)
```

```python
""" Esta tabela mostra o nivel de equilibrio de vida dos funcionarios demitidos e dos não demitidos por departamento
e a porcentagem de cada departamento por nivel de equilibrio de vida. Neste caso podemos observar uma alta porcentagem
nas notas mais altas sem nemhum indicio de problemas relacionado ao baixo percentual de notas
no geral a maior porcentagem esta nas notas 3."""

dep_equili = df.groupby(['Demissão','Equilíbrio vida trabalho','Departamento'])[['Departamento']]
dep_equili = df.groupby(['Demissão', 'Equilíbrio vida trabalho', 'Departamento']).size().reset_index(name='Total')

total_por_categoria = dep_satisf.groupby(['Demissão', 'Departamento'])['Total'].transform('sum')
dep_equili['Porcentagem'] = (dep_satisf['Total'] / total_por_categoria) * 100

display(dep_equili)
```

```python
''' Esta tabela mostra a media salarial de todos os funcionarios por cargo,
podemos observar neste caso que o grupo de Recursos humanos são os com menores salarios medios,
o que pode explicar a insatisfação dos funcionarios, e o alto indice de demissão do setor,
mesmo sendo um setor com menor numero de funcionarios.
'''
basededados = df.groupby('Cargo')[['Salário Mensal']].describe()
display(basededados)
```

```python
'''Esta tabela mostra o total de anos trabalhados dos funcionarios que foram demitidos, maior porcentagem esta presente
entre os 1 a 5 anos de empresa, o que pode indicar que os funcionarios com menor tempo de casa estão saindo mais rapidamente
da empresa do que se esperava.
'''
df2 = df.loc[:,["Demissão", "Nível de Satisfação com o ambiente de trabalho", "Satisfação com o trabalho", "Satisfação nas relações Não trabalho", "percentual de aumento de salário"]]
demitidos = df.loc[df['Demissão'] == 'Sim']
freq = demitidos['Total de anos trabalhados na empresa'].value_counts()
percentual = demitidos['Total de anos trabalhados na empresa'].value_counts(normalize=True)*100
anostrabalho = pd.DataFrame({'Frequência Absoluta':freq,'Frequência Percentual':percentual})
anostrabalho = anostrabalho.sort_values(by='Frequência Percentual')
anostrabalho.reset_index(inplace=True)



anostrabalho['Frequência Acumulada'] = anostrabalho['Frequência Absoluta'].cumsum()
anostrabalho['Frequência Acumulada %'] = (anostrabalho['Frequência Acumulada']/anostrabalho['Frequência Absoluta'].sum())*100

anostrabalho.loc[27] = ['TOTAL',anostrabalho['Frequência Absoluta'].sum(),
                  anostrabalho['Frequência Percentual'].sum(),'-','-']
display(anostrabalho)
```

```python
'''Esta tabela mostra a faixa dos anos trabalhados dos funcionarios demitidos e dos funcionarios não demitidos da empresa.
Analisando a tabela podemos ver uma porcentagem muito alta das demissões dos funcionarios que trabalharam poucos anos.
O que pode indicar que a empresa tende a manter os funcionarios com mais experiencia de trabalho.
'''

faixas = ['0 - 5',' 6 - 10','11 - 15','16 - 20','21 - 25','26 - 30','31 -40']
df['Faixa Anos trabalhados'] = pd.cut(x=df['Total de anos trabalhados'],bins=[0,5,10,15,20,25,30,40],labels=faixas)

dep_anostrab = df.loc[:,["Demissão", "Faixa Anos trabalhados", "Cargo", "Departamento", ]]

dep_anostrab = df.groupby(['Demissão','Faixa Anos trabalhados'])[['Departamento']]
dep_anostrab = df.groupby(['Demissão', 'Faixa Anos trabalhados']).size().reset_index(name='Total')

total_por_categoria = dep_anostrab.groupby(['Demissão'])['Total'].transform('sum')
dep_anostrab['Porcentagem'] = (dep_anostrab['Total'] / total_por_categoria) * 100

display(dep_anostrab)
```

```python
'''Esta tabela mostra a faixa das empresas trabalhadas dos funcionarios demitidos e dos funcionarios não demitidos da empresa,
Analisando a tabela podemos ver uma porcentagem muito alta das demissões dos funcionarios que trabalharam em poucas empresas,
Mas correlacionando os resultados com os grupos de demitidos e de não demitidos, podemos ver que a porcentagem é bem parecida,
Então as empresas trabalhadas anteriormente podem não ter nemhuma relação para explicar os casos de demissões dos funcionarios.
'''

faixas = ['0 - 2',' 3 - 5','6 - 8','9 - 11']
df['Empresas trabalhadas'] = pd.cut(x=df['qtde de empresas que já trabalhou'],bins=[0,2,5,8,11],labels=faixas)

dep_empretrab = df.loc[:,["Demissão", "Empresas trabalhadas", "Cargo", "Departamento", ]]

dep_empretrab = df.groupby(['Demissão','Empresas trabalhadas'])[['Departamento']]
dep_empretrab = df.groupby(['Demissão', 'Empresas trabalhadas']).size().reset_index(name='Total')

total_por_categoria = dep_empretrab.groupby(['Demissão'])['Total'].transform('sum')
dep_empretrab['Porcentagem'] = (dep_empretrab['Total'] / total_por_categoria) * 100

display(dep_empretrab)
```
