```python
# Listas criadas para salvar o funcionário no sistema

indice = []
nums_i = [1 , 2 , 3, 4, 5]
nomes_f = ['LUID', 'ALICE', 'LILIAN', 'GERALDO', 'MARCELLY']
cidades = ['IPATINGA', 'BRAGANCA', 'SÃO PAULO', 'FORTALEZA','BRASILIA']
estados = ['MG', 'SP', 'SP', 'CE', 'GO']
idades = [32, 17, 35, 38, 23]
escolaridades = ['CES', 'EMI', 'CES', 'ESF', 'EMI']
cargos = ['PL', 'PL', 'PL', 'PL','JR']
salarios = ['2000', '3000', '2000', '5000','1000']
emails = ['LUID@GMAIL', 'ALICE@GMAIL', 'LILIAN@GMAIL', 'GERALDO@GMAIL', 'MARCELLY@GMAIL']

# 1- Função para adicionar um novo funcionário ao sistema
def adicionar_funcionario():
    num_i = int(input("Digite o id : "))
    nome_f = input("Digite o nome do funcionário: ")
    cidade = input("Digite a cidade do funcionário: ")
    estado = input("Digite o estado do funcionário: ")
    idade = int(input("Digite a idade do funcionário: "))
    escolaridade = input('''=======================================================
EFC - ensino fundamental completo
EFI - ensino fundamental incompleto
EMC - ensino médio completo
EMI - ensino médio incompleto
CES - cursando ensino superior
ESF - ensino superior finalizado
OO - outro/ não informado
Digite a escolaridade do funcionário: ''')
    cargo = input('''=======================================================
AP - aprendiz
JR- júnior
PL - pleno
SR - sênior
SU - supervisor
GE - gerente
Digite o cargo do funcionário: ''')
    salario = float(input("Digite o salário do funcionário: "))
    email = input("Digite o e-mail do funcionário: ")

# 2- Armazenar as informações do funcionário nas listas

    nums_i.append(num_i)
    nomes_f.append(nome_f)
    cidades.append(cidade)
    estados.append(estado)
    idades.append(idade)
    escolaridades.append(escolaridade)
    cargos.append(cargo)
    salarios.append(salario)
    emails.append(email)
    print('''=======================================================
Funcionário adicionado com sucesso.''')

# 3- Função para atualizar as informações de um funcionário existente

def atualizar_funcionario():
    num_i = int(input('''=================Atualizar_Funcionário====================
Digite o ID do funcionário que deseja atualizar:'''))
    if num_i in nums_i:
        print('''=======================================================
Funcionário encontrado.''')
        indice = nums_i.index(num_i)
        nomes_f[indice] = input('''=======================================================
Digite o novo nome do funcionário: ''')
        cidades[indice] = input("Digite a nova cidade do funcionário:")
        estados[indice] = input("Digite o novo estado do funcionário:")
        idades[indice] = int(input("Digite a nova idade do funcionário: "))
        escolaridades[indice] = input('''=======================================================
EFC - ensino fundamental completo
EFI - ensino fundamental incompleto
EMC - ensino médio completo
EMI - ensino médio incompleto
CES - cursando ensino superior
ESF - ensino superior finalizado
OO - outro/ não informado
Digite a nova escolaridade do funcionário: ''')
        cargos[indice] = input('''=======================================================
AP - aprendiz
JR- júnior
PL - pleno
SR - sênior
SU - supervisor
GE - gerente
Digite o novo cargo do funcionário: ''')
        salarios[indice] = float(input("Digite o novo salário do funcionário: "))
        emails[indice] = input("Digite o novo e-mail do funcionário: ")
        print('''=======================================================
Informações do funcionário atualizadas com sucesso.''')
    else:
        print('''=======================================================
Funcionário não encontrado.''')

# 4- Função para exibir as informações de um funcionário específico

def exibir_funcionario():
    num_i = int(input("Digite o ID do funcionário que deseja visualizar: "))
    if num_i in nums_i:
        indice = nums_i.index(num_i)
        print('''=======================================================
ID:''', nums_i[indice])
        print("Nome:", nomes_f[indice])
        print("Cidade:", cidades[indice])
        print("Estado:", estados[indice])
        print("Idade:", idades[indice])
        print("Escolaridade:", escolaridades[indice])
        print("Cargo:", cargos[indice])
        print("Salário:", salarios[indice])
        print("E-mail:", emails[indice])
    else:
        print('''=======================================================
Funcionário não encontrado.''')

# 5- Função para listar todos os funcionários cadastrados

def listar_funcionarios():
    if nums_i:
        for num_i in range(len(nums_i)):
            print('''=======================================================
ID:''', nums_i[num_i])
            print("Nome:", nomes_f[num_i])
            print("Cidade:", cidades[num_i])
            print("Estado:", estados[num_i])
            print("Idade:", idades[num_i])
            print("Escolaridade:", escolaridades[num_i])
            print("Cargo:", cargos[num_i])
            print("Salário:", salarios[num_i])
            print("E-mail:", emails[num_i])
    else:
        print('''=======================================================
Nenhum funcionário cadastrado.''')

# 6- Função para remover um funcionário do sistema

def remover_funcionario():
    num_i = int(input("Digite o ID do funcionário que deseja remover: "))
    if num_i in nums_i:
        indice = nums_i.index(num_i)
        nums_i.pop(indice)
        nomes_f.pop(indice)
        cidades.pop(indice)
        estados.pop(indice)
        idades.pop(indice)
        escolaridades.pop(indice)
        cargos.pop(indice)
        salarios.pop(indice)
        emails.pop(indice)
        print('''=======================================================
Funcionário removido com sucesso.''')
    else:
        print('''=======================================================
Funcionário não encontrado.''')

# 7- Função menu para acesso as opções do sistema

while True:
   print('''=================Sitemas_De_Cadastro====================
Adicionar Funcionário: 1
Atualizar: 2
Exibir um funcionário: 3
Listar todos funcionários: 4
Excluir funcionário : 5
Sair: 6
=======================================================''')

   x = input('escolha um numero acima:')

   if x == '1':
    adicionar_funcionario()
   elif x == '2':
    atualizar_funcionario()
   elif x == '3':
     exibir_funcionario()
   elif x == '4':
     listar_funcionarios()
   elif x == '5':
      remover_funcionario()
   elif x == '6':
      print('sair do programa...')
      break
   else:
    print('Valor Invalido')
    print('=======================================================')
```
